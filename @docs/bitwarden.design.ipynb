# Bitwarden Crypto API

- to get the content, need to (D)Encrypt the ciphers.
- to access the ciphers, need the access token
- to get the access token, api client id and secret (from bitwarden server -> setup onetime)
- or
- to get the access token, need the refresh token
- to get the refresh token, need master key hash, optional 2fa key (from user TOTP generator)
- to get the master key hash, need the master key
- to (D)Encrypt the ciphers, need the unproteced symmetric key
- to get the unproteced symmetric key, need the protected symmetric key and the master key
- to get the protected symmetric key, need the access token
- to get the master key, need the master password (from user brain)
